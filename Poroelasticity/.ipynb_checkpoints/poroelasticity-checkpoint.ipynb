## __Problema PDE__

### Las ecuaciones de la mecánica del suelo poroso elástico
Las ecuaciones forman un sistema de ecuaciones para  la presión $p$ y los desplazamientos del medio $\boldsymbol{u}_s$:

\begin{align}
  \nabla \cdot \boldsymbol{\sigma}_{e} - \alpha\nabla p = \boldsymbol{f} \\
  s \frac{\partial p}{\partial t} + \alpha \frac{\nabla \cdot \boldsymbol{u}_f}{\partial t}+ \nabla \cdot \frac{k}{\mu} \nabla p=\boldsymbol{q}.
\end{align}

### Balance de Momento Lineal
\begin{align}
  \nabla \cdot \boldsymbol{\sigma}_{e} - \alpha\nabla p = \boldsymbol{f} \\
\end{align}

El lado derecho $f$ es una fuerza dada por unidad de volumen y al igual que las ecuaciones de elasticidad lineal,
$\sigma_e(\boldsymbol{u}_s)$ denota el tensor de tensión o esfuerzos efectivos, que para un sólido elástico está dado por

\begin{equation}
  \sigma_e(\boldsymbol{u}_s) = \lambda \text{tr}(\boldsymbol{\epsilon}(\boldsymbol{u}_s)) +2\mu\boldsymbol{\epsilon}(\boldsymbol{u}_s)  \boldsymbol{I},
\end{equation}

Los parámetros $\lambda$ y $\mu$ son los parámetros de Lamé respectivamente. Tenga en cuenta que hay dos ecuaciones de conservación de la cantidad de movimiento lineal: la del medio poroso elástico y la del fluido. La diferencia está en el términos adicional $\nabla p$ relativo a la presión dentro del medio elástico, así como de los términos transitorios en la del fluido. Los dos términos adicionales equilibran las fuerzas $F = \nabla\cdot\boldsymbol{\sigma} + \boldsymbol{f}$ por unidad de volumen en la segunda ley del movimiento de Newton.

Donde $\boldsymbol{\epsilon}(\boldsymbol{u})$ es el tensor de deformación

\begin{equation} \boldsymbol{\epsilon}(\boldsymbol{u}_s) = \frac{1}{2}\left(\nabla \boldsymbol{u}_s + (\nabla \boldsymbol{u}_s)^T\right).\end{equation}

### Balance de Masa
Balance de masa
Balance de la fase líquida
$$
\begin{aligned}
\begin{gathered}
\frac{\partial \phi \rho_f}{\partial t} + \nabla \cdot (\phi \rho_f v_f)=0\\
 \beta_{f}=-\frac{1}{V_f}\frac{\Delta V_f}{\Delta p} \ \ \ \ \ \ \mid \ \ \ \ \ \ \ \  \frac{\partial V_f}{V_f} = -\frac{\partial \rho_f}{\rho_f}\\
 \frac{\partial \rho f}{\partial t}=\rho_f \beta \frac{\partial p}{\partial t}\\
\mid \\
\frac{\partial \phi}{\partial t} +\frac{\partial p}{\partial t} \rho_f \beta + \nabla \cdot (\phi v_f)=0\\
\mid\\
\pm  \nabla \cdot\left(\phi v_s\right) \\
\mid\\
\frac{\partial \phi}{\partial t}+\phi \beta_f \frac{\partial p}{\partial t}+\nabla \cdot\left[\phi\left(v_f-v_s\right)\right]+\nabla \cdot\left(\phi v_s\right)=0 \\
\end{gathered}
\end{aligned}
$$
Balance de la fase sólida
$$
\begin{gathered}
\frac{\partial(1-\phi) \rho_s}{\partial t}+\nabla \cdot\left((1-\phi) \rho_s v_s\right)=0 \\
\mid \\
-\frac{\partial \phi}{\partial t}+\left(-\beta_s \frac{\partial \sigma}{\partial t}-\phi \beta_m \frac{\partial p}{\partial t}\right)+\nabla \cdot v_s=\nabla \cdot\left(\phi v_s\right) \\
\mid \\
\nabla \cdot v_s+\nabla \cdot\left[\phi\left(v_f-v_s\right)\right]+\phi\left(\beta+\beta_m\right) \frac{\partial p}{\partial t}-\beta \frac{\partial \sigma}{\partial t}=0\\
\mid \\
\end{gathered}
$$
$$
\boxed{\alpha \frac{\partial \nabla \cdot u}{\partial t}+S_\epsilon \frac{\partial p}{\partial t}+\nabla \cdot q=0}
$$

Ley de darcy
$$
q=-\frac{\kappa}{\mu} \nabla p
$$
$$
\boxed{\alpha \frac{\partial \nabla \cdot u}{\partial t}+S_\epsilon \frac{\partial p}{\partial t}-\nabla \cdot\left(\frac{\kappa}{\mu} \nabla p\right)=0}
$$

Algunos otros parámetros físicos incluidos en el modelo son los siguientes:


*   $\alpha = 1-\beta_s/\beta_m $ número de BIOT.

*   $\beta_s =  $ compresibidad del solido.

*   $\beta_m =$ compresibidad del medio poroso.

*   $s = (\alpha -\phi)\beta_s +\beta_f \phi $  es el coeficiente inercial.

*   $\phi$  porosidad del medio.

*   $\beta_f $ compresibidad del fluido.

## __Formulación variacional__
Si se considera primero una partición de elementos finitos $\mathcal{T}_h =\left\{K\right\}$ del dominio $\Omega$. El diámetro de la partición de elementos se denota por $h$.
se define los espacios de elementos finitos $\boldsymbol{V}_h$ y $W_h$ construidos a partir de funciones polinómicas continuas a trozos en el espacio.

La versión completamente discretizada del problema de Galerin se puede escribir entonces como

encontrar un $p_h^{j+1} \in W_h $ y un $\boldsymbol{u}_h^{j+1} \in \boldsymbol{V}_h$, tal que satisfaga




\begin{align}
     \int_{\Omega} \nabla \boldsymbol{\sigma}_e(\boldsymbol{u}_h^{j+1}) : \nabla \boldsymbol{v}_h dx - \int_{\Omega} \alpha p_h^{j+1} (\nabla \cdot \boldsymbol{v}_h) dx = \int_{\Omega} \boldsymbol{f}\cdot\boldsymbol{v}_h\: dx +\int_{\partial\Omega} (\boldsymbol{n} \cdot \boldsymbol{\sigma})\cdot\boldsymbol{v}_h \: ds
\end{align}
\begin{align}
 \int_{\Omega} s \frac{\partial p_h^{j+1}}{\partial t} w_h\: dx + \int_{\Omega} \alpha \frac{\partial(\nabla \cdot  \boldsymbol{u})_h^{j+1}}{\partial t}w_h\:dx + \int_{\Omega} \nabla w_h \cdot \frac{k}{\mu} \nabla p_h^{j+1} \: dx = -\int_{\partial\Omega} \frac{k}{\mu}p_h^{j+1} n w_h\; ds
\end{align}

### Esquemas de discretización temporal: Backward Differential Formula
Esquema BDF de primer orden:
\begin{align}
    \frac{\partial \phi_h^{j+1}}{\partial t} = \frac{\phi_h^{j+1}-\phi_h^j}{\delta t}+O(\delta t)
\end{align}
Esquema BDF de segundo orden:
\begin{align}
    \frac{\partial \phi_h^{j+1}}{\partial t}=\frac{3 \phi_h^{j+1}-4 \phi_h^j+\phi_h^{j-1}}{2 \delta t}+O\left(\delta t^2\right)
\end{align}
Esquena BDF de tercer orden:
\begin{align}
    \frac{\partial \phi_h^{j+1}}{\partial t}=\frac{11 \phi_h^{j+1}-18 \phi_h^j+9 \phi_h^{j-1}-2 \phi_h^{j-2}}{6 \delta t}+O\left(\delta t^3\right)
\end{align}

## Consolidación de Terzaghi
Simulamos el ensayo de consolidación de suelos o caso de terzaghi con el fin de validar con una solucion analitica nuestra aproximación. En la consolidación de terzaghi tenemos una muestra de material sometida a una carga constante con un drenaje en la misma frontera de la carga, con lo que se espera que la carga inicial que soporta la presión de poros se disipe en el tiempo.

Importamos las librerias necesarias

```python
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
from dolfin import *
from matplotlib import pyplot
import numpy as np
import mshr
import math
```

definimos la malla en el dominio $\Omega$ y sus fronteras

```python
H=1
D=0.5
mesh=RectangleMesh(Point(-float(D/2), 0.0), Point(float(D/2),-H), int(D*10), abs(int(H*10)),"crossed")

contorno = MeshFunction("size_t", mesh, mesh.topology().dim()-1)
tol =1E-3
class load(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and abs(x[1])< tol
class bound(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and abs(x[1]+H)< tol
class walls(SubDomain):
        def inside(self, x, on_boundary):
            return on_boundary and ((abs(x[0]+D/2)< tol) or (abs(x[0]-D/2)< tol) )
load().mark(contorno, 2)
bound().mark(contorno, 5)
walls().mark(contorno, 1)
```

Relaciones constitutivas y relaciones cinematicas

```python
#deformacion
def epsilon(u):
    return 0.5*(grad(u) + grad(u).T)

#esfuerzo
def sigma(u):
    return lmbda*div(u)*Identity(d) + 2*mu*epsilon(u)
```

Definición de los espacios funcionales

```python
ele_p  = FiniteElement("P",mesh.ufl_cell(), 1) # pressure
ele_u  = VectorElement("P",mesh.ufl_cell(), 1) # solid displacement
W = MixedElement([ele_p, ele_u])
W = FunctionSpace(mesh, W)
U = TrialFunction(W)
V = TestFunction(W)
Z= FunctionSpace(mesh, 'P', 1)
Z_d= FunctionSpace(mesh, 'DG', 0)
Z_v = VectorFunctionSpace(mesh, 'P', 1)
TS = TensorFunctionSpace(mesh, "P", 1)


p, u = split(U)
q, v = split(V)
```

Propiedades físicas

```python
B_f=4.4E-10
Poro=0.3
nu=0.3#coeficiente de poisson
flo=Constant((0,0))
E=50000000# #modulo elasticidad
mu = E/(2*(1+nu))#coeficientes de Lame
lmbda = E*nu/((1+nu)*(1-2*nu))#coeficientes de Lame
B_m=(E/(3*(1-2*nu)))**(-1)
B_s=B_m/10
alpha=(1-B_s/B_m) #biotcoef
s_coef=(alpha-Poro)*B_s +Poro*B_f
theta =18.94#K(subd,18.94,20.61,23.27,20.53,21.84) #angulos friccion interna
C=15530#K(subd,15530,10350,18650,18400,14000) #cohesion
K=5E-7
kappa=K
mv=1/((1/B_m)+((4/3)*mu))
cv=K/(alpha**2*mv+s_coef)
carga=5E7
```

Pasos de tiempo

```python
t=0
dtdot=0.001
delta= dtdot/(cv/abs(H)**2)
steps=int( 2/dtdot)+10
dt=Constant((delta))
```

Condiciones de Contorno

```python
bp1=DirichletBC(W.sub(0),Constant((0)),contorno,2)

bc1 = DirichletBC(W.sub(1).sub(0),(0.0),contorno,1)
bc2 = DirichletBC(W.sub(1), Constant((0.0,0.0)),contorno,5)
bcs=[bc1,bc2,bp1]
```

Formulación Varacional

```python
d = u.geometric_dimension()
f = Constant((0, 0))
n=FacetNormal(mesh)#vector normal
x = SpatialCoordinate(mesh)


x_n=Expression(('0','0','0'),degree=3)
X_n = Function(W)
X_n=interpolate(x_n, W)
p_n, u_n = split(X_n)

X_nn=Function(W)
X_nn=interpolate(x_n, W)
p_nn, u_nn = split(X_nn)

X_nnn=Function(W)
X_nnn=interpolate(x_n, W)
p_nnn, u_nnn = split(X_nnn)

# esquemas de discretización temporal
#pconst=[3./2,-2,1./2,0.0] #bdf2
#pconst = [0.48*11/6+0.52*3/2,0.48*-3+0.52*-2,0.48*3/2+0.52*1/2,0.48*-1/3] #bdf2 op
pconst= [11/6,-3,3/2,-1/3] #bdf 3
#pconst=[1,-1,0,0] #bdf1
types=['BDF1','BDF2','BDF2op','BDF3']
scheme=types[3]
du=pconst[0]*u
du_n=pconst[1]*u_n
du_nn=pconst[2]*u_nn
du_nnn=pconst[3]*u_nnn
du_t= du+du_n +du_nn +du_nnn

divu=pconst[0]*div(u)
divu_n=pconst[1]*div(u_n)
divu_nn=pconst[2]*div(u_nn)
divu_nnn=pconst[3]*div(u_nnn)
divu_t= divu+divu_n +divu_nn+divu_nnn

dp=pconst[0]*p
dp_n=pconst[1]*p_n
dp_nn=pconst[2]*p_nn
dp_nnn=pconst[3]*p_nnn
dp_t=dp+dp_n+dp_nn+dp_nnn


ds = Measure('ds', domain=mesh, subdomain_data=contorno)

T=Constant((0,-carga))

#_____________ Ecuacion de conservacion del momentum lineal
F1 =  inner(sigma(u), epsilon(v))*dx -alpha*p*div(v)*dx\
    -inner(T, v)*ds(subdomain_id=2, domain=mesh, subdomain_data=contorno)

#_________________ecuacion de conservacion de masa

F2 = dt*inner(nabla_grad(q), K*nabla_grad(p))*dx \
    + alpha*divu_t*q*dx + s_coef*(dp_t)*q*dx\
    -dt*inner(flo,n)*q*ds(subdomain_id=5,domain=mesh, subdomain_data=contorno)  -dt*inner(flo,n)*q*ds(subdomain_id=1,domain=mesh, subdomain_data=contorno)
L_momentum =lhs(F1)
R_momentum =rhs(F1)
L_mass=lhs(F2)
R_mass=rhs(F2)
L=L_momentum+L_mass
R=R_momentum+R_mass


```

Solución analítica

```python

snaps=100

p0= ((alpha*mv)/(alpha**2*mv+s_coef))*carga
def p_analitico(time,snaps,cv,p0):
    z=0
    for n in range(snaps):
        for i in range(500):
                if i==0:
                    p=0
                else:
                    c=math.sin((2*i-1)*math.pi*(z)/(2*H))
                    b=math.exp(-(((2*i-1)*math.pi)/(2*H))**2*cv*time)
                    d=((1)/(2*i-1))
                    p=p+d*c*b
        p=4/math.pi*(p)
        if n==0:
            a=np.array([[p,z]])
        else:
            a= np.append(a,np.array([[p,z]]),axis=0)
        z=z+1/snaps
    return a
def u_analitico(time,snaps,cv,p0):
    c=-mv*carga*H
    for i in range(500):
            if i==0:
                p=0
            else:
                b=math.exp(-(((2*i-1)*math.pi)/(2*H))**2*cv*time)
                d=((1)/(2*i-1)**2)
                p+=b*d
    a=8*alpha*mv*H/(math.pi**2)*p0*(p)+c
    return a
```

```python
a = p_analitico(1E-5,snaps,cv,p0)
plt.plot(a)
```

```python


def sigma_1 (T): #Primer esfuerzo principal
    c=det(T)
    b =-tr(T)
    return -b/2 + sqrt(b**2-4*c)/2

def sigma_3(T): #Primer esfuerzo principal
    c=det(T)
    b =-tr(T)
    return -b/2 - sqrt(b**2-4*c)/2


X_w = Function(W)

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)
L2=[]
uplot=[]
SvsP=[]
u_f=mv*carga
print('u final',u_f)
ig, ax = plt.subplots()
dtdot=(cv/1**2)*delta
t=delta
for pot in range(steps):


    solve(L==R,X_w,bcs)
    X_nnn.assign(X_nn)
    p_nnn, u_nnn = split(X_nnn)
    X_nn.assign(X_n)
    p_nn, u_nn = split(X_nn)
    X_n.assign(X_w)
    p_n, u_n = split(X_n)
    u_=project(u_n,Z_v)
    p_=project(p_n,Z)

    # post proceso
    if pot==0:
        u_0dot = (mv-(alpha**2*mv**2)/(alpha**2*mv+s_coef))*carga
        p_0=carga#p_(0.00,-0.1)
    tdot=(cv/H**2)*t
    s = sigma(u_)
    cauchy=project(s,TS)

    #o1 = sigma_1(cauchy)

    #o2 = sigma_3(cauchy)

    #tm=(o1-o2)/2 #mean stress
    principal=project(cauchy[1,1],Z)
    SvsP.append([p_(0,-H)/1000,-principal(0,-H)/1000,tdot])
    uplot.append([(u_analitico(t, snaps, cv, p0)+u_0dot)/(u_f-u_0dot),(u_(0,0)[1]+u_0dot)/(u_f-u_0dot),tdot])
    z_=0
    for k in range(snaps):
        pdot=p_(0.0,z_*1)/p0
        if k ==0:
            results=np.array([[pdot,-z_]])
        else:
            results =np.append(results,np.array([[pdot,-z_]]),axis=0)
        z_=z_- 1/snaps
    p_a = p_analitico(t,snaps,cv,p0)
    L2.append([np.sum((results[:,0]-p_a[:,0])**2),tdot])
    if near(t,0.01/(cv/H**2),dt/2) or near(t,0.1/(cv/H**2),dt/2) or near(t,0.5/(cv/H**2),dt/2)or near(t,1/(cv/H**2),dt/2):


        line1, =ax.plot(results[:, 0], results[:, 1], "-",color='red',label='FEM',)
        plt.xlabel("$P*$ ",fontsize=20)
        plt.ylabel("$z*$",fontsize=20)
        line2,=ax.plot(p_a[:, 0], p_a[:, 1], "--",color='black',label='Analítica')

        lines = plt.gca().get_lines()
        l1=lines[-1]
        if near(t,0.01/(cv/1**2),dt/2) :
            ax.legend(handler_map={line1: HandlerLine2D(numpoints=4)},loc= 'upper center')

        plt.ylim((0,1.05))
        plt.xlim((0,1.05))
    if near(t,2/(cv/H**2),dt/2):
        break
    print('u max:',u_.vector().get_local().min(),'step', pot, 'of', steps,'time*:',tdot)
    t += delta
plt.show()
```
