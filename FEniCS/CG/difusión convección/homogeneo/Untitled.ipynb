```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import sys
import math

types=['BDF1','BDF2','exp']

error=pd.DataFrame()
it = 0
ax = plt.figure(figsize=(20,10)).add_subplot(projection='3d')
#ax.set_yscale('log')
#ax.set_xscale('log')
#ax.set_zscale('log')
for scheme in types:
    df1 = pd.read_csv(f'time_error_{scheme}.csv')
    error=df1.rename(columns={"L2_norm_sum": scheme})            
    ax.plot(np.log(error["cpu_time"]),np.log(error['dt']),np.log(error[scheme]),label=scheme, marker='o')
    print(error)
ax.set_xlabel("cpu time [s] ",fontsize=15)
ax.set_zlabel("$L_\infty$ norm",fontsize=15)
ax.set_ylabel("dt",fontsize=15)
ax.set_xlim(0, 10)
ax.set_ylim(-20, 1)
ax.set_zlim(-100, 10)
ax.legend(loc= 'upper right')
ax.view_init(elev=25, azim=245, roll=0)
plt.show()
```
